```python
df['bore']=df['bore'].replace('?',df['bore'].mean())
df['stroke']=df['stroke'].replace('?',df['stroke'].median())
df['horespower']=df['horsepower'].replace('?',df['horsepower'].median())
df['peak-rpm']=df['peak-rpm'].replace('?',df['peak-rpm'].median())
df['price']=df['price'].replace('?',df['price'].median())
```
